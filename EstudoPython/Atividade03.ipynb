```python
""" 
Formação Cientista de Dados - Prof. Fernando Amaral
SEÇÃO 13: Prática em Python - Gráficos, Visualização e Dashboards

1. Arquivo DadosS12-13.csv
    1. CODIGO
    2. MUNICIPIO
    3. PIB
    4. VALOREMPENHO
    
"""
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as srn
```

```python
base = pd.read_csv('../Dados/DadosS12-13.csv', sep=';')
base.shape
```

```python
""" 
2. Empenho em diárias em municípios gaúchos e PIB 
do municipio.

"""
srn.histplot(base['VALOREMPENHO']).set_title('Empenho')
```

```python
agrupado = base.sort_values('PIB').head(10)
agrupado = agrupado.iloc[:,1:3]
agrupado
agrupado.plot.bar(x='MUNICIPIO',y='PIB', color = 'blue')
```

```python
""" 
3. Execute um processo de Análise de Dados 
Exploratória para conhecimento/Entendimento 
deste dados.   

"""

agrupado = base.sort_values('VALOREMPENHO').head(10)
agrupado = agrupado.iloc[:,[1,3]]
agrupado
agrupado.plot.bar(x='MUNICIPIO',y='VALOREMPENHO', color = 'gray')
```
